## **Data Visualization**

#### **1. Defining functions for constructing missing value heatmaps, numeric distributions and categorical counts**

```python
## Function for missing value heatmaps
def plot_missing_heatmap (df, title="Missing values Heatmap") :
    if df.isnull().sum().sum() > 0 :
        msno.heatmap(df)
        plt.title(title, fontsize=14, weight="bold")
        plt.show()
    else:
        print("No missing values present in the dataset")
        
```

```python
## Function for numeric distribution
def plot_numeric_dist (df, col, bins=30):
    plt.figure(figsize=(6,4))
    sns.histplot(df[col],bins=bins,kde=True)
    plt.title(f"Distribution of {col}", fontsize=12, weight="bold")
    plt.show()
```

```python
## Function for categorical counts
def plot_categorical_counts (df, col):
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f"Counts of {col}", fontsize=12, weight="bold")
    plt.xticks(rotation=45)
    plt.show() 
```

```python
# Time series / datetime distribution
def plot_datetime_distribution(df, column, bins=30):
    plt.figure(figsize=(10,5))
    sns.histplot(df[column], bins=bins)
    plt.title(f'Distribution over time: {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()
```

```python
# Top N categories by numeric column
def plot_top_n(df, category_col, numeric_col, n=10):
    top_n = df.groupby(category_col)[numeric_col].mean().sort_values(ascending=False).head(n)
    plt.figure(figsize=(10,5))
    sns.barplot(x=top_n.index, y=top_n.values)
    plt.title(f'Top {n} {category_col} by average {numeric_col}')
    plt.xticks(rotation=45)
    plt.show()
```

#### **2. Using functions to create visualization of each dataset one at a time**

##### **A. Accounts Dataset**

```python
#Missing value heatmap
plot_missing_heatmap(accounts,title="Accounts dataset missing values")

#Numeric distribution for "seats" column
plot_numeric_dist(accounts,'seats')

#Categorical distributions
categorical_cols = ['industry', 'country', 'plan_tier', 'referral_source', 'is_trial', 'churn_flag']
for col in categorical_cols:
    plot_categorical_counts(accounts, col)

#Signup trends over time
accounts['signup_date'] = pd.to_datetime(accounts['signup_date'],dayfirst=True)
plot_datetime_distribution(accounts, 'signup_date')
```

##### **B. Subscriptions Dataset**

```python
#Missing value heatmap
plot_missing_heatmap(subscriptions,title="Missing values in Subscriptions dataset")

#Numerical Distribution
numeric_cols = ['seats', 'mrr_amount', 'arr_amount']
for col in numeric_cols:
    plot_numeric_dist(subscriptions, col)

#Categorical Count
categorical_cols = ['plan_tier', 'billing_frequency', 'is_trial', 
                    'upgrade_flag', 'downgrade_flag', 'churn_flag', 'auto_renew_flag']
for col in categorical_cols:
    plot_categorical_counts(subscriptions, col)

#Time based visualization
# Convert dates to datetime
subscriptions['start_date'] = pd.to_datetime(subscriptions['start_date'],dayfirst=True)
subscriptions['end_date'] = pd.to_datetime(subscriptions['end_date'], errors='coerce',dayfirst=True)  # handle missing end_dates

# Subscriptions started over time
plot_datetime_distribution(subscriptions, 'start_date')

# Distribution of subscription end dates (only for non-null)
plot_datetime_distribution(subscriptions[subscriptions['end_date'].notnull()], 'end_date', bins=50)

# Churn by Plan Tier
plt.figure(figsize=(8,5))
sns.countplot(x='plan_tier', hue='churn_flag', data=subscriptions)
plt.title("Churn by Plan Tier", fontsize=14, weight='bold')
plt.show()

# MRR Distribution by Plan Tier
plt.figure(figsize=(8,5))
sns.boxplot(x='plan_tier', y='mrr_amount', data=subscriptions)
plt.title("MRR Distribution by Plan Tier", fontsize=14, weight='bold')
plt.show()
```

##### **C. Feature_usage Dataset**

```python
#Missing values heatmap
plot_missing_heatmap(feature_usage, title="Missing Values in Feature Usage Dataset")

# 2. Numeric distributions
numeric_cols = ['usage_count', 'usage_duration_secs', 'error_count']
for col in numeric_cols:
    plot_numeric_dist(feature_usage, col)

# 3. Categorical counts
categorical_cols = ['feature_name', 'is_beta_feature']
for col in categorical_cols:
    plot_categorical_counts(feature_usage, col)

#Convert dates to date time
feature_usage['usage_date'] = pd.to_datetime(feature_usage['usage_date'], dayfirst=True)

#Datetime distribution
plot_datetime_distribution(feature_usage, 'usage_date', bins=50)

# 5. Top N features by numeric metrics
plot_top_n(feature_usage, category_col='feature_name', numeric_col='usage_count', n=10)
plot_top_n(feature_usage, category_col='feature_name', numeric_col='usage_duration_secs', n=10)
plot_top_n(feature_usage, category_col='feature_name', numeric_col='error_count', n=10)

# 6. Beta vs Non-Beta Features Comparison
beta_usage = feature_usage.groupby('is_beta_feature')[['usage_count','usage_duration_secs','error_count']].mean().reset_index()
beta_usage_melted = beta_usage.melt(id_vars='is_beta_feature', var_name='Metric', value_name='Average')
plt.figure(figsize=(10,5))
sns.barplot(x='Metric', y='Average', hue='is_beta_feature', data=beta_usage_melted)
plt.title("Beta vs Non-Beta Feature Usage")
plt.show()
```

##### **D. Support_tickets Dataset**

```python
#Missing value heatmap
plot_missing_heatmap(support_tickets, title="Missing Values Heatmap for Support Tickets")

#Numeric Distribution
numeric_columns = ['resolution_time_hours', 'first_response_time_minutes', 'satisfaction_score']
for col in numeric_columns:
    plot_numeric_dist(support_tickets, col)

#Categorical count
categorical_columns = ['priority', 'escalation_flag']
for col in categorical_columns:
    plot_categorical_counts(support_tickets, col)

# Convert submitted_at and closed_at to datetime first
support_tickets['submitted_at'] = pd.to_datetime(support_tickets['submitted_at'],dayfirst=True)
support_tickets['closed_at'] = pd.to_datetime(support_tickets['closed_at'],dayfirst=True)

#Time-series plot
plot_datetime_distribution(support_tickets, 'submitted_at', bins=30)
plot_datetime_distribution(support_tickets, 'closed_at', bins=30)

#Top N accounts by number of tickets
def plot_top_accounts(df, n=10):
    top_accounts = df['account_id'].value_counts().head(n)
    plt.figure(figsize=(10,5))
    sns.barplot(x=top_accounts.index, y=top_accounts.values)
    plt.title(f'Top {n} Accounts by Number of Tickets')
    plt.xticks(rotation=45)
    plt.ylabel('Number of Tickets')
    plt.show()

plot_top_accounts(support_tickets, n=10)

#Average resolution time by priority
plot_top_n(support_tickets, category_col='priority', numeric_col='resolution_time_hours', n=3)

#Average first response time by priority
plot_top_n(support_tickets, category_col='priority', numeric_col='first_response_time_minutes', n=3)

#Satisfaction score distribution by priority (custom violin plot)
plt.figure(figsize=(10,5))
sns.violinplot(x='priority', y='satisfaction_score', data=support_tickets)
plt.title('Satisfaction Score Distribution by Priority')
plt.show()

#Escalation vs resolution time (custom boxplot)
plt.figure(figsize=(10,5))
sns.boxplot(x='escalation_flag', y='resolution_time_hours', data=support_tickets)
plt.title('Resolution Time by Escalation Status')
plt.show()
```

##### **E. Churn_events Dataset**

```python
#Missing Values heatmap
plot_missing_heatmap(churn_events, title='Missing Values in Churn Event Dataset')

#Numeric distribution
plot_numeric_dist(churn_events, 'refund_amount_usd')

#Categorical Counts
plot_categorical_counts(churn_events, 'reason_code')
plot_categorical_counts(churn_events, 'preceding_upgrade_flag')
plot_categorical_counts(churn_events, 'preceding_downgrade_flag')
plot_categorical_counts(churn_events, 'is_reactivation')

# Convert date column to datetime
churn_events['churn_date'] = pd.to_datetime(churn_events['churn_date'], dayfirst=True)

#Time Distribution
plot_datetime_distribution(churn_events, 'churn_date', bins=30)

#Top N Categories by Numeric Column
plot_top_n(churn_events, 'reason_code', 'refund_amount_usd', n=10)

# Count of missing vs non-missing feedback
feedback_counts = churn_events['feedback_text'].notnull().value_counts()
plt.figure(figsize=(6,4))
sns.barplot(x=feedback_counts.index, y=feedback_counts.values)
plt.xticks([0,1], ['Missing', 'Provided'])
plt.title('Feedback Text Availability')
plt.ylabel('Count')
plt.show()
```
